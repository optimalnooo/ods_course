```python
# подгружаем все нужные пакеты
import pandas as pd
import numpy as np

# игнорируем warnings
import warnings
warnings.filterwarnings("ignore")

import seaborn as sns

import matplotlib
import matplotlib.pyplot as plt
import matplotlib.ticker
%matplotlib inline

# настройка внешнего вида графиков в seaborn
sns.set_context(
    "notebook", 
    font_scale = 1.,       
    rc = { 
        "figure.figsize" : (12, 9), 
        "axes.titlesize" : 18 
    }
)
```

```python
train = pd.read_csv('../data/mlbootcamp5_train.csv', sep=';',
                    index_col='id')
```

```python
print('Размер датасета: ', train.shape)
train.head()
```

```python
# 1.
cols_of_interest = ['height', 'gender', 'weight', 'cholesterol', 'gluc', 'smoke', 'alco']
sns.heatmap(train[cols_of_interest].corr(method='pearson'), annot=True, linewidths=0.5);
```

```python
# 2.
estimate_gender_df = pd.melt(train, value_vars=['height'], id_vars='gender')
sns.violinplot(x='variable', y='value', hue='gender', data=estimate_gender_df, scale='count');
```

```python
sns.kdeplot(data=train[train['gender']==1]['height']);
sns.kdeplot(data=train[train['gender']==2]['height']);
```

```python
# 3.
cols_of_interest = ['height', 'weight', 'age', 'ap_hi', 'ap_lo', 'cholesterol', 'gluc', 'cardio', 'smoke', 'alco']
sns.heatmap(train[cols_of_interest].corr(method='spearman'), annot=True, linewidths=0.5);
```

```python
# 4.
filtered_ap = train[(train['ap_lo'] > 0) & (train['ap_hi'] > 0)]

g = sns.jointplot(x='ap_lo', y='ap_hi', data=filtered_ap);
ax = g.ax_joint
ax.set_xscale('log')
ax.set_yscale('log')
```

```python
# 5.
train['age_years'] = (train['age'] // 365.25).astype(int)
sns.countplot(x='age_years', hue='cardio', data=train);
```
